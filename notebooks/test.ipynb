```python
import pandas as pd
df = pd.read_csv("../data/dataset.csv", index_col=0)
# Check for duplicates in 'track_name' only
df
```

```python
features = ['track_id', 'artists', 'album_name', 'track_name', 'popularity', 'duration_ms', 'explicit',
                         'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness',
                         'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature', 'track_genre']

agg_dict = {}
for col in features:
    if col == 'popularity':
        agg_dict[col] = 'mean'
    elif col == 'track_genre':
        agg_dict[col] = lambda x: list(x.unique())
    else:
        agg_dict[col] = 'first'

# print(agg_dict)

combined_df = df.groupby('track_id', as_index=False).agg(agg_dict)
```

```python
combined_df['artists'] = combined_df['artists'].str.split(';')
combined_df
```
